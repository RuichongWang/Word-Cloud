```python
import re
import matplotlib.pyplot as plt
import numpy as np
import requests
from bs4 import BeautifulSoup
from PIL import Image
from wordcloud import WordCloud

# we first go to wikipedia to get some passages for run the script.
res = requests.get('https://en.wikipedia.org/wiki/Google')
res = BeautifulSoup(res.text)
res = res.find_all('div', attrs={'class': 'mw-parser-output'})
words = res[0].get_text().strip()

print(words[:200]) # Let's see what does our sample passage looks like
```

```python
# choose the shape
mask = np.array(Image.open(("heart.jpg")))

wc = WordCloud(max_words=500,
               background_color='white',
               mask=mask[:, :, 1],
               margin=10).generate(words)

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.show()
```

```python
wc.to_file('Google!.pdf')
```
